```python
import pandas as pd
```

```python
elections = pd.read_csv('elections.csv')
# return a Series, where the index is the "Party", and the value is the number of times that party won an election
party_counts = elections['Party'].value_counts()
# return a Series, where the index is the "Year" and whose value are the number of candidates that ran in that year
candidate_counts = elections['Year'].value_counts()
candidate_counts
```

```python
# return a DataFrame, which keeps only the parties that have at least one election "%" more than 50%. There is column called "%"
filtered_parties = elections.sort_values("%", ascending=False).groupby('Party').agg(lambda x: x.iloc[0])
# filtered_parties = elections.groupby('Party').agg("min")
filtered_parties
```
